```python
import util, lines, group
```

```python
# path to pdf file with three identation types
file_path = "../test_files/LS_index_1920.pdf"

# read pdf with fitz
pdf_words, pdf_dicts = util.read_pdf(file_path) 
```

```python
# make lines data frame
words_df = lines.make_words_df(pdf_words, start_page=1)
lines_df = words_df.rename(columns={"text": "line_text"})
lines_df["x0"] = [round(x, 2) for x in lines_df["x0"]]
lines_df["y0"] = [round(x, 2) for x in lines_df["y0"]]
lines_df["x1"] = [round(x, 2) for x in lines_df["x1"]]
lines_df["y1"] = [round(x, 2) for x in lines_df["y1"]]
lines_df = lines.merge_close_lines(lines_df)
lines_df = lines.remove_useless_lines(lines_df)

lines_df
```

```python
# filter lines from one specified page for demonstration purposes
page = 3
df = lines_df.loc[lines_df["page"]==page]

df
```

```python
# group lines based on their x0-coordinate
x0_groups = group.group_rows(df, by="x0", mode="fitz")

# sort groups by mean x0 value
x0_groups = x0_groups.sort_values(by="last_x0_mean")

x0_groups
```

```python
# get lines in first group -> countries
df.loc[x0_groups.iloc[0]["lines"]]
```

```python
# get lines in second group -> start of entries
df.loc[x0_groups.iloc[1]["lines"]]
```
