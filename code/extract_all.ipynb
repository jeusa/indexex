```python
mode = "tess"

if mode=="tess":
    files = tess_dfs
elif mode=="fitz":
    files = indexes_files

print(f"Mode: {mode} \n")

for i, f in enumerate(files):
    print(f"Starting index extraction for {f}")
    
    f_name, f_suffix = os.path.splitext(f)
    save_path = output_dir + os.path.basename(f_name) + f"_{mode}.csv"
    
    if mode=="tess":
        extract.extract_indexes_tess(f, save_to=save_path)
    elif mode=="fitz":
        extract.extract_indexes_pdf(f, save_to=save_path)
    
    print(f"Done with file {i+1}/{len(files)}")
```
